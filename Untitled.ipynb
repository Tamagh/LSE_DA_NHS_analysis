```python
# Libraries and settings needed for analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size.
sns.set(rc={'figure.figsize':(15, 12)})

# Set the plot style as white.
sns.set_style('white')

# Maximum column width to display
pd.options.display.max_colwidth = 200
```

```python
tweets = pd.read_csv('tweets.csv')
tweets.head
```

```python
tweets_text = tweets[['tweet_full_text']]
tweets_text.head
```

```python
tags = []
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
print(tags[:30])
```

```python
tags = pd.Series(tags)
tags
```

```python
print(tags[:30])
```

```python
tags.value_counts()
```

```python
data = pd.DataFrame(tags.value_counts().reset_index().values, columns=['word','count'])
```

```python
print(data.dtypes)
```

```python
data.head(30)
```

```python
data.info()
```

```python
data['count'] = data['count'].astype(int)
```

```python
data.info()
```

```python
data.describe()
```

```python
ax = sns.barplot(x='word', y='count', data=data[data['count']>10])
```

```python
data_filtered = data.iloc[2:]
data_filtered.head()
```

```python
ax = sns.barplot(x='count', y='word', data=data_filtered[data_filtered['count']>11], palette='light:b_r')
ax.set_title('Top trending on Twitter')
ax.set(xlabel='Hashtag counts')
ax.set(ylabel='')
plt.tight_layout()
sns.set(font_scale=1.5)
sns.set_style('white')
```
